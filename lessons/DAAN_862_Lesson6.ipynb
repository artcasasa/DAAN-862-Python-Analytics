```python
# 1. Plot am-based histogram to compare mpg
plt.figure(figsize=(10, 6))  # Set the figure size
plt.hist(mtcars[mtcars['am'] == 1]['mpg'], alpha=0.7, label='Manual (am=1)', bins=10, color='blue')  # Histogram for manual
plt.hist(mtcars[mtcars['am'] == 0]['mpg'], alpha=0.7, label='Automatic (am=0)', bins=10, color='orange')  # Histogram for automatic
plt.xlabel('Miles Per Gallon (mpg)')  # X-axis label
plt.ylabel('Frequency')  # Y-axis label
plt.title('Comparison of MPG based on Transmission Type (am)')  # Plot title
plt.legend()  # Show the legend
plt.show()  # Display the plot
```

```python
# 2. Scatterplot of mpg VS. hp
plt.figure(figsize=(10, 6))  # Set the figure size
plt.scatter(mtcars['hp'], mtcars['mpg'], c='green')  # Scatter plot with horsepower vs. mpg
plt.xlabel('Horsepower (hp)')  # X-axis label
plt.ylabel('Miles Per Gallon (mpg)')  # Y-axis label
plt.title('Scatterplot of MPG vs HP')  # Plot title
plt.show()  # Display the plot
```

```python
# 3. Scatterplot matrix for new data consisting of columns [disp, hp, drat, wt, qsec]
plt.figure(figsize=(12, 12))  # Set the figure size for scatter matrix
scatter_matrix(mtcars[['disp', 'hp', 'drat', 'wt', 'qsec']], alpha=0.8, diagonal='hist', marker='o', color='purple', s=30)
plt.suptitle('Scatterplot Matrix for disp, hp, drat, wt, qsec')  # Add a title to the entire figure
plt.show()  # Display the plot
```

```python
# 4. Boxplots for new data consisting of columns [disp, hp, drat, wt, qsec]
plt.figure(figsize=(12, 8))  # Set the figure size for the boxplot
plt.boxplot([mtcars['disp'], mtcars['hp'], mtcars['drat'], mtcars['wt'], mtcars['qsec']], 
            labels=['disp', 'hp', 'drat', 'wt', 'qsec'])
plt.title('Boxplots for disp, hp, drat, wt, qsec')  # Plot title
plt.ylabel('Values')  # Y-axis label
plt.xlabel('Variables')  # X-axis label
plt.show()  # Display the plot
```

```python
# 5. Additional scatterplots to evaluate the impact of various variables on mpg
fig, axs = plt.subplots(2, 2, figsize=(14, 10))  # Create a 2x2 grid of subplots

# Scatterplot of mpg vs. disp
axs[0, 0].scatter(mtcars['disp'], mtcars['mpg'], color='red')
axs[0, 0].set_xlabel('Displacement (disp)')
axs[0, 0].set_ylabel('Miles Per Gallon (mpg)')
axs[0, 0].set_title('MPG vs Displacement')

# Scatterplot of mpg vs. hp
axs[0, 1].scatter(mtcars['hp'], mtcars['mpg'], color='blue')
axs[0, 1].set_xlabel('Horsepower (hp)')
axs[0, 1].set_ylabel('Miles Per Gallon (mpg)')
axs[0, 1].set_title('MPG vs Horsepower')

# Scatterplot of mpg vs. wt
axs[1, 0].scatter(mtcars['wt'], mtcars['mpg'], color='green')
axs[1, 0].set_xlabel('Weight (wt)')
axs[1, 0].set_ylabel('Miles Per Gallon (mpg)')
axs[1, 0].set_title('MPG vs Weight')

# Scatterplot of mpg vs. qsec
axs[1, 1].scatter(mtcars['qsec'], mtcars['mpg'], color='purple')
axs[1, 1].set_xlabel('Quarter Mile Time (qsec)')
axs[1, 1].set_ylabel('Miles Per Gallon (mpg)')
axs[1, 1].set_title('MPG vs Quarter Mile Time')

plt.tight_layout()  # Adjust layout to avoid overlap
plt.show()  # Display the plot
```
